```python
import numpy as np
import pandas as pd

def convert_to_days(value):
    # If value is NaN, return NaN
    if pd.isna(value):
        return np.nan

    value = str(value).lower().replace('+', '').strip()  # convert to string & normalize

    if "today" in value:
        return 0

    if "yesterday" in value:
        return 1

    if "hour" in value:
        return 0

    if "week" in value:
        num = int(value.split()[0])
        return num * 7

    if "day" in value:
        num = int(value.split()[0])
        return num

    return np.nan

df['Posted_Days_Ago'] = df['Date Posted'].apply(convert_to_days)
```

```python
import re

def clean_location(val):
    if pd.isna(val):
        return val

    # remove bracket content like (All Areas)
    val = re.sub(r"\(.*?\)", "", val)

    # split by comma or dash
    parts = re.split(r"[-,]", val)
    parts = [p.strip() for p in parts if p.strip()]

    # if starts with hybrid or remote, take last city name
    if parts[0].lower() in ["hybrid", "remote", "wfh", "work from home"]:
        city = parts[-1]
    else:
        city = parts[0]

    return city

# apply extraction
df['Location'] = df['Location'].apply(clean_location)

# merge duplicate names (no NCR grouping!)
df['Location'] = df['Location'].replace({
    'Gurgaon/ Gurugram': 'Gurugram',
    'Delhi / NCR': 'Delhi',
    'Bengaluru': 'Bangalore',
    'Bangalore Rural': 'Bangalore'
})

# Remove the word Remote if you don't want it at all
# df = df[df['Location'].str.lower() != 'remote']

# Reset index after cleanup
df.reset_index(drop=True, inplace=True)
```
